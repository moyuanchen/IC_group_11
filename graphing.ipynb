```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plotting unhedged PnL
cost = 1000
fee = 1350
fx = np.linspace(1.1, 1.5, 100)
pnl = 1350 - cost * fx

#let's say we hedge our currency risk with a forward contract with delivery fx = 1.28
fx_hedged = 1.28 * np.ones(100)
pnl_hedged = 1350 - cost * fx_hedged

#option strike = 1.28, premium = 0.05
# the option will be excercised if the fx is higher than 1.28
pnl_option = 1350 - np.array([cost * x if x < 1.28 else cost * 1.28 for x in fx]) - cost * 0.05
#pnl_option = 


plt.plot(fx, pnl, label='Unhedged PnL')
plt.plot(fx, pnl_hedged, label='forward hedged PnL')
plt.plot(fx, pnl_option, label='option hedged PnL')
plt.xlabel('FX rate USD/EUR')
plt.ylabel('Operating Profit')
plt.legend()
plt.grid()
plt.savefig('per_student_hedging.png')
```

```python
%matplotlib qt
def unhedged_pnl_per(fx, cost=1000, fee=1350):
    return fee - cost * fx

def forward_hedged_pnl_per(fx, cost=1000, fee=1350, fx_hedged=1.28):
    return fee - cost * fx_hedged

def option_hedged_pnl_per(fx, cost=1000, fee=1350, strike=1.28, premium=0.05):
    return fee - np.array([cost * x if x < strike else cost * strike for x in fx]) - cost * premium

def unhedge_pnl(fx, delta_V, cost=1000, fee=1350):
    return fee - cost * fx

def forward_hedged_pnl(fx, e_V, cost=1000, fee=1350, fx_hedged=1.28):
    return fee - fx_hedged * cost * (e_V + 1)+ fx * cost * e_V

#3D plot forward hedged pnl as function of fx and e_V
fx = np.linspace(1.1, 1.5, 100)
e_V = np.linspace(-0.25, 0.25, 100)
fx, e_V = np.meshgrid(fx, e_V)
Z = forward_hedged_pnl(fx, e_V)
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(fx, e_V, Z)
ax.set_xlabel('fx')
ax.set_ylabel('e_V')
ax.set_zlabel('PnL')
#ax.view_init(elev=30, azim=45)
#plt.savefig('forward_hedged_pnl.png')

```
